Optical beat signals are created when the optical field consists of more than one frequency component. Lets look at the simplest case of just two frequency components: at some arbitrary point in space we have an optical field given by

\begin{align}
E(t) = a_c \mathrm{e}^{\omega_c t} + a_1 \mathrm{e}^{\omega_1 t}.
\end{align}

Here, the subscript $c$ stands for carrier and the subscript $1$ indicates a sideband. We want to distinguish these as for a gravitational wave detector, the carrier frequency is around 282 THz ($\lambda = 1064$ nm) while an upper sideband frequency preferably can be written as $\omega_1 = \omega_c + \omega_s$, where $\omega_s$ is somewhere between $10\times 2\pi $ rad/s and $10^9$ rad/s.  As we use units where the amplitude of the light field is expressed in $\sqrt{\mathrm{W}}$, the power is given by

\begin{align}
\bar{P}(t)=E(t)E^*(t) &= \big|a_c\big|^2 + \big|a_1\big|^2 + a_c a_1^* \mathrm{e}^{-i\omega_s t} + a_c^* a_1 \mathrm{e}^{i\omega_s t} \\
                      &= \big|a_c\big|^2 + \big|a_1\big|^2 + 2\mathcal{Re} \big(a_c^* a_1 \mathrm{e}^{i\omega_s t} \big).
\end{align}

This is the red trace shown in the figure below, while the DC-power and the time dependent part are shown in black and blue, respectively. The grey trace shows the electrical field. So we end up with a DC part, and an AC part of frequency $\omega_s = \big| \omega_c - \omega_1 \big|$. We say that the two frequency components beat against each other to crate this **beat signal**. The sideband amplitude is generally tiny compared to the carrier amplitude, so in the DC part it can be neglected. However, the time dependent term is proportional to the sideband amplitude and can provide information about the mechanism that created the sideband.

We can generalise this to arbitrary many frequency components:

\begin{align}
\bar{P}(t) = \sum_j \sum_k a_j a_k^* \mathrm{e}^{i(\omega_j - \omega_k)t} ,
\end{align}


where the sums are over all frequency components in the field. For the field with three frequency components, created by the moving mirror, the output becomes

\begin{align}
\bar{P}(t) = \big| a_0 \big|^2 +  2\mathcal{Re}\big[ (a_0a_-^* + a_0^*a_+) \mathrm{e}^{i\Omega t}\big],
\end{align}

where we have omitted terms of second order in sideband amplitude since these are small compared to the carrier amplitude. However, if the field is phase modulated, the second term would be zero since phase fluctuations do not give rise to power flucations. And power is what a photodiode measures. What a Michelson interferometer does is to convert phase modulation into amplitude modulation. So no amplitude modulation, no beat signal.

**Summary:** Beat signals shows up in the power of optical fields consisting of two or more frequency components, where the beat frequency is the difference between the frequencies present in the field. The beat signal is an effect of the field being amplitude modulated.
